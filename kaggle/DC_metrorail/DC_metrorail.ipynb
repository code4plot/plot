## Visualizing foot traffic of DC metrorail stations - Part 1
The objective of this notebook is to visualize foot traffic of the DC metrorail stations at various times of the day, as described in the source of this dataset.

This dataset contains 4 columns:
1) Entering station
2) Exit station
3) Entering time period
4) Average weekday riders

The Time Period are defined as follows:
AM Peak = opening to 9:30am
Midday = 9:30am to 3:00pm
PM Peak = 3:00pm to 7:00pm
Evening = 7:00pm to midnight

```python
import pandas as pd
#load dataset
df = pd.read_csv("May-2012-Metrorail-OD-Table-by-Time-of-Day-and-Day-of-Week .csv")
df.head()
```

```python
#simplify column names
df.rename(columns = {'Ent Station':'enter', 'Ext Station':'exit', 'Ent Time Period':'period',
       'Riders, Average Weekday, May 2012':'avgrider'}, inplace = True)
```

Here, I wrote a custom function, "getFootTraffic", to compile foot traffic per station (enter + exit). If enter == exit, I count as one.

```python
#foot traffic per station
#if entry station = exit station, don't double count
from stations_utility import getFootTraffic
footTraffic = getFootTraffic(df)
footTraffic.head()
```

```python
footTraffic.shape
```

```python
#distribution of riders
footTraffic.plot.hist(bins = 15, figsize = (10,5), title = "Figure 1. Distribution of foot traffic")
```

```python
footTraffic.sort_values(by = 'rider', ascending = False).plot.bar(figsize = (20,5), title = "Figure 2. Distribution of Foot traffic per station")
```

```python
#there are different time periods
df.period.unique()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex=True)
n = 3
for i in ['AM Peak', 'Midday', 'PM Peak', 'Evening']:
    ft = getFootTraffic(df[df.period == i].reset_index())
    if n == 3:
        ft.sort_values(by = 'rider', ascending = False).plot.bar(figsize = (20,5),ax = axes[n])
        stationOrder = ft.sort_values(by = 'rider', ascending = False).index
    else:
        ft.reindex(stationOrder).plot.bar(figsize = (20,5), ax = axes[n])
    axes[n].set_ylabel(i)
    n-=1
axes[0].title.set_text("Figure 3. Distribution of foot traffic per station\n(sorted on AM Peak)")
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex=True)
n = 3
for i in ['AM Peak', 'Midday', 'PM Peak', 'Evening']:
    ft = df[df.period == i].groupby("enter").sum()
    if n == 3:
        ft.sort_values(by = 'avgrider', ascending = False).plot.bar(figsize = (20,5),ax = axes[n])
        axes[n].set_xlabel("")
        stationOrder = ft.sort_values(by = 'avgrider', ascending = False).index
    else:
        ft.reindex(stationOrder).plot.bar(figsize = (20,5), ax = axes[n])
    axes[n].set_ylabel(i)
    n-=1
axes[0].title.set_text("Figure 4. Distribution of foot traffic entering a station\n(sorted on AM Peak)")
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex=True)
n = 3
for i in ['AM Peak', 'Midday', 'PM Peak', 'Evening']:
    ft = df[df.period == i].groupby("exit").sum()
    if n == 3:
        ft.sort_values(by = 'avgrider', ascending = False).plot.bar(figsize = (20,5),ax = axes[n])
        axes[n].set_xlabel("")
        stationOrder = ft.sort_values(by = 'avgrider', ascending = False).index
    else:
        ft.reindex(stationOrder).plot.bar(figsize = (20,5), ax = axes[n])
    axes[n].set_ylabel(i)
    n-=1
axes[0].title.set_text("Figure 5. Distribution of foot traffic exiting a station\n(sorted on AM Peak)")
```

## foot traffic volume for each metro station evolves across different periods of time
Union Station enjoys the highest total foot traffic volume across the entire day (Figure 2). However, if one groups the dataset by the time period, Gallery Place - Chinatown takes over in the evening (Figure 3), presumably for retail activities (dinner, bar, etc).  Similarly in Figure 4 & 5, one would observe that many are entering/exiting Gallery Place - Chinatown in the evening.

One would also be interested to see how foot traffic changes across different geographical areas of Washington.  For example, Smithsonian station is located in the Washington Monument Grounds.  It enjoys little "entry' traffic in the AM Peak, relative to Midday and PM Peak (Figure 4).  Most people visit this place Midday (Figure 5).  One could also study the foot traffic of Reagan Washington National Airport, and perhaps correlate it with domestic flight schedule.  In any case, one could infer that most departures and arrivals occur in the Midday, as it enjoys the most foot traffic than all other times of the day.
